```python
dev = torch.device('cuda:1')
mwugan = MWUGAN(conf, dataset=dataset, device=dev, gan_batchsize=128, update_method='naive', delta=0.25)
for i in range(50):
    mwugan.iteration(dis_iters=2000, gan_iters=0)
    a=np.sum(mwugan.weights_list[-1])
    b=np.sum(mwugan.weights_list[-1][60000:])
    print(a,b)
    print(b/a)
```

```python
import os
base_dir = 'data/fashion_naive/'
for i in range(50):
    print('save:', i)
    path = base_dir+'gan{0:02d}'.format(i)

    if not os.path.exists(path):
        os.makedirs(path)
    for j in range(50):
        savename = path+'/batch{0:03d}.pt'.format(j)
        a = mwugan.gan_list[i].gen_samples(600)
        torch.save(a, savename)
    c_weight = mwugan.weights_list[i]
    savefile_weights = path+'/weights.npy'
    np.save(savefile_weights, c_weight)
```
